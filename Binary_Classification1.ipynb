# Binary Classification Problem

```python
from warnings import filterwarnings
filterwarnings('ignore')
```

### Step 1: Read the dataset

```python
import pandas as pd
url1 = "https://raw.githubusercontent.com/utkarshg1/Neural-Net-Datasets/main/booking_train.csv"
df = pd.read_csv(url1)
```

```python
df.head()
```

### Step 2: Perform basic data quality checks

```python
df.info()
```

```python
df.isna().sum()
```

```python
df.duplicated().sum()
```

### Step 3: Seperate X and Y(booking_status)
id is stastically insignificant

```python
X = df.drop(columns = ["id", "booking_status"])
Y = df[["booking_status"]]
```

```python
X.head()
```

```python
Y.head()
```

### Step 4: Perform preprocessing on X

```python
X.dtypes
```

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
```

```python
num_pipe = Pipeline(steps=[("impute", SimpleImputer(strategy="median")),
                           ("scaler", StandardScaler())]).set_output(transform='pandas')
```

```python
X_pre = num_pipe.fit_transform(X)
X_pre.head()
```

### Step 5: Train test split

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(X_pre, Y,
                                                test_size=0.2, random_state=42)
```

```python
xtrain.head()
```

```python
ytrain.head()
```

```python
xtest.head()
```

```python
ytest.head()
```

```python
xtrain.shape
```

```python
xtest.shape
```

### Building a neural network

image.png

```python
xtrain.shape[1]
```

### Dropout layer - Randomly drop some % of neuron links

```python
from keras.models import Sequential
from keras.layers import Input, Dense, Dropout
```

```python
# Initialize the model
model = Sequential()

# Input Layer
model.add(Input(shape=(xtrain.shape[1], )))

# Hidden Layer 1
model.add(Dense(64, activation='relu'))
# Add Dropout layer 1
model.add(Dropout(0.4))

# Hidden Layer 2
model.add(Dense(32, activation='relu'))
# Add Dropout layer 2
model.add(Dropout(0.3))

# Add output layer
model.add(Dense(1, activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(loss='binary_crossentropy',
              metrics=["accuracy"],
              optimizer='adam')
```

```python
hist = model.fit(xtrain, ytrain, validation_split=0.2, epochs=60)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 6))
plt.plot(hist.history['loss'], label='Training loss')
plt.plot(hist.history['val_loss'], label='Validation Loss')
plt.xlabel("Epochs")
plt.ylabel("Binary Cross entropy Loss")
plt.title("Learning Curve")
plt.legend()
plt.show()
```

```python
model.evaluate(xtrain, ytrain)
```

```python
model.evaluate(xtest, ytest)
```

### Get predictions on train and test

```python
yprob_train = model.predict(xtrain)
yprob_test = model.predict(xtest)
```

```python
ypred_train = [1 if prob>=0.5 else 0 for prob in yprob_train]
ypred_test = [1 if prob>=0.5 else 0 for prob in yprob_test]
```

### Confusion Matrix

```python
from sklearn.metrics import ConfusionMatrixDisplay
ConfusionMatrixDisplay.from_predictions(ytest, ypred_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(ytest, ypred_test))
```

```python
from sklearn.metrics import roc_auc_score
roc_auc_score(ytest, yprob_test)
```
